### Actual HW#3-2 Solution

```python
avg_temp_dict = dict()
for i in range(1,32):
    response = urllib.request.urlopen(f"https://www.metaweather.com/api/location/2487956/2019/8/{i}")
    data = response.read()
    weather_dict = json.loads(data.decode("utf-8"))
    
    datestring = f"2019-08-{i:02d}"
    temps = [ entry['the_temp'] for entry in weather_dict if entry['created'].startswith(datestring)]
    avg = sum(temps)/len(temps)
    print(datestring, temps, avg)
    avg_temp_dict[datestring] = avg
    
print(avg_temp_dict)
```
