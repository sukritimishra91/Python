### Adjusting the Plot: Line Colors and Styles

```python
plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');
```

```python
# For short, you can use the following codes:
plt.plot(x, x + 0, linestyle='-')  # solid
plt.plot(x, x + 1, linestyle='--') # dashed
plt.plot(x, x + 2, linestyle='-.') # dashdot
plt.plot(x, x + 3, linestyle=':');  # dotted
```

These single-character color codes reflect the standard abbreviations in the RGB (Red/Green/Blue) and CMYK (Cyan/Magenta/Yellow/blacK) color systems, commonly used for digital color graphics.

```python
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r');  # dotted red
```

### Adjusting the Plot: Axes Limits

```python
plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
```

```python
plt.plot(x, np.sin(x))

# axis displayed in reverse
plt.xlim(10, 0)   
plt.ylim(1.2, -1.2);
```

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5]);
```

```python
plt.plot(x, np.sin(x))
plt.axis('tight');
```

### Labeling Plots

```python
plt.plot(x, np.sin(x))

# add figure title, x and y labels
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)"); #by using semicolon we do not get the location in memory storage(<matplotlib.legend.Legend at 0x11c260ef0>)
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')

plt.axis('equal')  # aspect ratio = 1

plt.legend();  # add legend
```

# Scatter Plots

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.scatter(x, y, marker='o');
```

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
y2 = np.sin(x) + 0.5
y3 = np.sin(x) + 1

plt.scatter(x, y, marker='o', color='blue');
plt.scatter(x, y2, marker='+', color='red');
plt.scatter(x, y3, marker='x', color='green');
```

For scatter plots, we can control the following:

* alpha: transparency level
* cmap: color map (<a href="https://matplotlib.org/1.4.1/users/colormaps.html">see doc</a>, <a href="https://bids.github.io/colormap/">more colormaps</a>)
* colorbar

```python
plt.cm   # tab to show available colormaps
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, 
            c=colors, s=sizes, 
            alpha=0.3,
            cmap='jet')
plt.colorbar();  # show color scale
```

# Contour Plots

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)  # create 2D grids from 1D arrays
Z = f(X, Y)
```

```python
plt.contour(X, Y, Z);
```

```python
plt.contour(X, Y, Z, 20, cmap='plasma');  # specify # of lines & coloramps
```

#### Filled contour plot

```python
plt.contourf(X, Y, Z, 20, cmap='magma');
```

```python
plt.contourf(X, Y, Z, 20, cmap='viridis');
plt.colorbar()
```

# Histograms

```python
data = np.random.randn(1000)

plt.hist(data)
```

```python
counts, bin_edges = np.histogram(data, bins=5)
print(counts)
```

The hist() function has many options to tune both the calculation and the display:

```python
plt.hist(data, 
         bins=30, 
         alpha=0.5,
         histtype='stepfilled', 
         color='steelblue',
         edgecolor='none');
```

# Subplots

```python
# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
```

```python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)

# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```

# 2D Histograms

```python
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T  #T transpose 
```

```python
plt.hist2d(x, y, bins=30, cmap='plasma')
cb = plt.colorbar()
cb.set_label('counts in bin')  # labels in the colorbar
```

# 3D Plots

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot(x, y, z)
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

# Make data.
x = np.outer(np.linspace(-2, 2, 30), np.ones(30))
y = x.copy().T # transpose
z = np.cos(x ** 2 + y ** 2)

# Plot the surface.
surf = ax.plot_surface(x, y, z, cmap='jet', # edgecolor='none',
                       linewidth=0, antialiased=False)

fig.colorbar(surf, shrink=0.5, aspect=5)
```
