```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

### Set Seaborn

```python
import seaborn as sns
sns.set()
```

```python
data_np = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
```

```python
# put it in dataframe
data = pd.DataFrame(data_np, columns=['x', 'y'])
```

```python
data
```

```python
plt.hist(data['x'], alpha=0.3)
plt.hist(data['y'], alpha=0.3)
```

# Univariate & bivariate kernel density estimate plots

<a href="https://seaborn.pydata.org/generated/seaborn.kdeplot.html">seaborn.kdeplot() doc</a>

```python
for col in 'xy':
    sns.kdeplot(data[col], shade=True)
```

```python
sns.kdeplot(data['x'], data['y']);
```

# Univariate distribution plots

<a href="https://seaborn.pydata.org/generated/seaborn.distplot.html">seaborn.distplot() doc</a>

```python
for col in 'xy':
    sns.distplot(data[col])
```

# Plots of two variables with bivariate and univariate graphs

<a href="https://seaborn.pydata.org/generated/seaborn.jointplot.html">seaborn.jointplot() doc</a>

```python
with sns.axes_style('white'):
    sns.jointplot("x", "y", data, kind='kde');
```

```python
with sns.axes_style('white'):
    sns.jointplot("x", "y", data, kind='hex')
```

# Categorical Plots

<a href="https://seaborn.pydata.org/generated/seaborn.catplot.html">seaborn.catplot() doc</a>

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
with sns.axes_style(style='ticks'):
    g = sns.catplot("day", "total_bill", "sex", data=tips, kind="box")
    g.set_axis_labels("Day", "Total Bill");
```

# Bar Plots

```python
planets = sns.load_dataset('planets')
planets.head()
```

```python
with sns.axes_style('white'):
    g = sns.factorplot("year", data=planets, aspect=4.0, kind='count',
                       hue='method', order=range(2001, 2015))
    g.set_ylabels('Number of Planets Discovered')
```

# More Plots


* <a href="https://seaborn.pydata.org/generated/seaborn.PairGrid.html">Pair Grids</a>

* <a href="https://seaborn.pydata.org/generated/seaborn.violinplot.html">Violin Plots</a>

```python
data = pd.read_csv("marathon-data.csv")
data.head()
```

### Convert time strings to seconds

```python
data['split_sec'] = pd.to_timedelta(data['split']).dt.total_seconds()
data['final_sec'] = pd.to_timedelta(data['final']).dt.total_seconds()
data['split_frac'] = 1 - 2 * data['split_sec'] / data['final_sec']

data.head()
```

### Pair grid of scatter plots

```python
g = sns.PairGrid(data, vars=['age', 'split_sec', 'final_sec', 'split_frac'],
                 hue='gender', palette='RdBu_r')
g.map(plt.scatter, alpha=0.8)
g.add_legend();
```

### Violin Plots

```python
sns.violinplot("gender", "split_frac", data=data,
               palette=["lightblue", "lightpink"]);
```
